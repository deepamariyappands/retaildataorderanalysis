```python
cursor.execute("""
    CREATE TABLE if not exists df_orders1 (
        order_id INT,
        order_date DATE,
        ship_mode VARCHAR(20),
        segment VARCHAR(20),
        country VARCHAR(20),
        city VARCHAR(20),
        state VARCHAR(20),
        postal_code VARCHAR(20),
        region VARCHAR(20),
        category VARCHAR(20),
        sub_category VARCHAR(20)
    )
""")
```

```python
df1.columns
```

```python
query = """
     INSERT INTO df_orders1(order_id, order_date, ship_mode, segment, country, city,
       state, postal_code, region, category , sub_category
       )
       VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""       
for i,row in df1.iterrows():
    values = (row['order_id'],row['order_date'],row['ship_mode'],row['segment'],row['country'],row['city'],row['state'],row['postal_code'],
              row['region'],row['category'],row['sub_category'])
    cursor.execute(query,values)
    myconnection.commit()
    
```
